```python
import numpy as np
import pandas as pd
```

```python
dict1 = {
    "Name" : ["Hassan" , "Ali" , "Umar" , "Abdulrehman"],
    "Gender" : ["Male", "Male", "Male" , "Khusra"],
    "Age" : [19,23,20,18]
}
# 1. How to make dataframe
df  = pd.DataFrame(dict1)

# 2. how to make csv to that dataframe
df.to_csv("CSV")

# 3. To delete the indexing of that csv file you can write:
df.to_csv("CSV_WithoutIndex",index=False)

# 4. If the dataframe is too big then:
df.head(1) # Only show first head
df.tail(1) # Only show last tail

# 5. To only show the numerical rows or columns in the dataframe with different operations
df.describe()
```

```python
# Making new Dataframe
newdf = pd.DataFrame(np.random.rand(334,5),np.arange(334)) # make a dataframe containing 334 rows and 5 columns
newdf.describe() # shows different operations on newdf
type(newdf) # See the type of newdf

# 2. Shows columns
newdf.columns

# 3. Shows all the indexes
newdf.index

# 4. Take transpose of newdf
newdf.T

# 5. To sort newdf
# with axis= 0 ,ascending=False(sort form bottom to up)
# with axis = 1,ascending=False(reverse left to right)
# axis = 0 (row)
# axis = 1(column)
newdf.sort_index(axis=1,ascending=False)

# 5. To change the value of specific cell in the dataFrame:
# dont use this newdf[0][0] = "Hassan" as this is  not a good practice
newdf.loc[0,0] = "Hassan"
newdf

# 6. Change the index of columns
newdf.columns = np.array(["A","B","C","D","E"])
newdf

# OR

newdf.columns = list("ABCDE")
newdf

# 7. To drop a specific column or row
newdf.drop("A",axis=1) # columns "A"
newdf.drop(0,axis = 0) # row = 0
newdf

# 8. Some other uses of loc
newdf.loc[[1,2],["A","B"]]  # give [1,2] row and ["A","B"] column
newdf.loc[:,["A","B"]] # give all rows and ["A","B"]
newdf.loc[[0,1],:] # Give [0,1] rows and all columns

# 9. Check using conditions
newdf.loc[(newdf["B"]<0.3)]
newdf.loc[(newdf["B"]<0.3) & (newdf["C"] < 0.5)]

# 10. to completely remove columns or rows form dataframe or modify dataframe
# we use inplace for that
newdf.drop(["A","B"] , axis = 1,inplace=True)
newdf

# 11. To again start indexing or reset indexing after modifying
newdf.reset_index(drop=True,inplace=True)
newdf

# 12. to drop duplicates form dataframe
newdata = pd.DataFrame(np.ones((3,4))).astype(int)
newdata = newdata.drop_duplicates(subset=[1])
newdata

# 13. To show null values in specific row or column
newdf["C"].isnull()

# 14. to see where is null and where is not in dataframe
# to see where is not null
newdf.notnull()

# to see where is null
newdf.isnull()
```
